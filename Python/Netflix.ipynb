### Data Visualization


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Total number of Indian TV Shows released every year

```python
a = data[(data['type'] == 'TV Show') & (data['country'] == 'India')]
plt.figure(figsize=(10,5))
sns.countplot(x='release_year',data=a,order = a.groupby(by=['release_year'])
              ['title'].count().index)
plt.title('Number of Indian TV Shows released every year')
plt.show()
```

What is the Percentage of Movies and TV Shows released World Wide

```python
labels = ['Movies','TV Shows']
plt.pie(data['type'].value_counts(),labels = labels,colors=['blueviolet','orange'],
explode = [.1,.1],autopct = '%.2f%%')
plt.title('Percentage of Movies and TV Shows released world wide')
plt.show()
```

Number of Movies and TV Shows released each year in India

```python
a = data[data['country']=='India']
plt.figure(figsize=(25,15))
sns.countplot(x='release_year',data=a,
              hue = 'type')
plt.xticks(rotation = 'vertical')
plt.title("Number of Movies and TV Shows released each year in India")
plt.show()
```

Number of TV Shows and Movies released every year World Wide

```python
plt.figure(figsize=(15,8))
sns.countplot(x='release_year',data=data,order=data.groupby(by=['release_year'])
              ['title'].count().index)
plt.title("Number of TV Shows and Movies released every year world wide")
plt.xticks(rotation='vertical')
plt.show()
```

Number of Movies Released in every Rating Category

```python
a = data[data['type']=='Movie']
plt.figure(figsize=(10,5))
sns.countplot(x='rating',data =a,order=a.groupby(by=['rating'])
             ['title'].count().index)
plt.title("Number of Movies released in every Rating Category")
plt.show()
```
